# Heart Diseases Data set consists of:

age: The person's age in years

sex: The person's sex (male,female)

cp: The chest pain experienced (Value 0: typical angina, Value 1: atypical angina, Value 2: non-anginal pain, Value 3: asymptomatic)

trestbps: The person's resting blood pressure (mm Hg on admission to the hospital)

chol: The person's cholesterol measurement in mg/dl

fbs: The person's fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false)

restecg: Resting electrocardiographic measurement (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria)

thalach: The person's maximum heart rate achieved

exang: Exercise induced angina (1 = yes; 0 = no)

oldpeak: ST depression induced by exercise relative to rest ('ST' relates to positions on the ECG plot. See more here)

slope: the slope of the peak exercise ST segment (Value 0: upsloping, Value 1: flat, Value 2: downsloping)

ca: The number of major vessels (0-4)

thal: A blood disorder called thalassemia

target: Heart disease (1= Heart Disease, 0= No heart Disease)

# importing column transformer

```python
from sklearn.compose import ColumnTransformer
```

# importing simple imputer

```python
df1.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
# imputing missing values
ct1=ColumnTransformer(
    transformers=[
        ('imp1', SimpleImputer(strategy='mean'),[3])], remainder='passthrough')
```

```python
d=ct1.fit_transform(df1)
```

```python
d
```

# catergorical encoding

```python
# oneHot encoder encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
ct2=ColumnTransformer([
    ('encoder',OneHotEncoder(sparse=False,drop='first'),[2])],remainder='passthrough')
```

```python
e=ct2.fit_transform(d)
e
```

```python
e.shape
```

#Standard scaling

```python
from sklearn.preprocessing import MinMaxScaler
# scaling dataset
ct3=ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,13))
])
```

```python
f=ct3.fit_transform(e)
f
```
